```python
import matplotlib.pyplot as plt
import numpy as np

# ================================
# Data
# ================================
models = ["ResNet-50", "ViT-S/16"]
shape_bias = [47.22, 51.88]
texture_bias = [52.78, 48.12]

x = np.arange(len(models))  # positions
width = 0.35  # bar width

# ================================
# Plot
# ================================
plt.figure(figsize=(8,6))
bars1 = plt.bar(x - width/2, shape_bias, width, label="Shape Bias", color="steelblue")
bars2 = plt.bar(x + width/2, texture_bias, width, label="Texture Bias", color="orange")

# Add values on top
for bars, values in [(bars1, shape_bias), (bars2, texture_bias)]:
    for bar, val in zip(bars, values):
        plt.text(bar.get_x() + bar.get_width()/2, val + 1, f"{val:.2f}%",
                 ha="center", va="bottom", fontsize=10)

# Formatting
plt.title("Shape vs Texture Bias (STL-10)", fontsize=14)
plt.ylabel("Percentage (%)", fontsize=12)
plt.xticks(x, models)
plt.ylim(0, 100)  # keep scale honest
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ["ResNet-50", "ViT-S/16"]
accuracy = [31.50, 95.40]
consistency = [20.22, 96.29]

x = np.arange(len(models))
width = 0.35

# Plot
fig, ax = plt.subplots(figsize=(8,6))
bars1 = ax.bar(x - width/2, accuracy, width, label="Accuracy", color="#4C72B0")
bars2 = ax.bar(x + width/2, consistency, width, label="Consistency under shifts", color="#55A868")

# Labels and title
ax.set_ylabel("Percentage (%)")
ax.set_title("Translation Invariance Test Results (STL-10)")
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.set_ylim(0, 110)
ax.legend()

# Annotate values
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f"{height:.2f}%",
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # offset
                    textcoords="offset points",
                    ha="center", va="bottom", fontsize=9)

plt.show()
```
